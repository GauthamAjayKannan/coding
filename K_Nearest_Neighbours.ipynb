```python
# Import LabelEncoder
from sklearn import preprocessing
import pandas as pd
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

dataset_dir='/Users/kiruthikasekar/Workspace/workshop/datasets/'
output_dir='/Users/kiruthikasekar/Workspace/workshop/output/'

dataset=pd.read_csv(dataset_dir+'weather_knn.csv')
dataset=dataset.drop(['Unnamed: 0'], axis=1)
dataset
```

```python
# knn doesn't work on categorical data so encode it to numerical data using labelencoder
#creating labelEncoder
le = preprocessing.LabelEncoder()
# Converting string labels into numbers.
dataset.weather=le.fit_transform(dataset.weather)
dataset.temp=le.fit_transform(dataset.temp)
dataset.play=le.fit_transform(dataset.play)
# After dataset encoding
dataset
```

```python
#split dataset in features and target variable
feature_cols = ['weather','temp']
X = dataset[feature_cols] # Features
y = dataset.play # Target variable
```

```python
# Split dataset into training set and test set
from sklearn.model_selection import train_test_split # Import train_test_split function
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0) # 70% training and 30% test
```

```python
#Import knearest neighbors Classifier model
from sklearn.neighbors import KNeighborsClassifier

#Create KNN Classifier
knn = KNeighborsClassifier(n_neighbors=3)

#Train the model using the training sets
knn.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = knn.predict(X_test)
```

```python
# Performance Measure
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

```
